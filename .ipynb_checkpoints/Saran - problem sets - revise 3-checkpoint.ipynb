## ðŸ‘‡ Questions 1 to 4 ask you to build, execute, and backtest a strategy based on one kind of momentum strategy called the `Moving Average Convergence Divergence` (MACD) crossover.

### <font color = blue> ðŸ‘‰ Question 1</font>. Letâ€™s take a look at Apple (AAPL) as our security of interest, over the three-year period of 2015 through 2017.

#### Moving Average Convergence Divergence (`MACD`) is a lagging, trend-following momentum indicator reflecting the relationship between two moving averages of stock prices.

#### The strategy utilizes two indicators, the MACD and the MACD signal line:

- The MACD is defined as the `difference` between the `12`-day ***exponential*** moving average and the `26`-day exponential moving average.
- The MACD signal line is then defined as the `9`-day ***exponential*** moving average of the MACD.

### <font color = green> NOTE: The MACD crossover strategy is defined as:

- A `bullish` crossover arises when the MACD line turns `upward and crosses` beyond the MACD signal line.
- A `bearish` crossover arises when the MACD line turns `downward and crosses` under the MACD signal line.

### Visualize your buy and sell positions.

### Below are the lines of code that lead to an answer:

```python
#1.) import data
start = '2015-01-01'
end = '2017-12-31'
ticker = 'AAPL'
data = pdr.get_data_yahoo(ticker, start = start, end = end)
```

```python
#2.) Calculate return and ema
price = 'Close'
df = data[[price]].copy()
df['return'] = df[price].pct_change()
df['ema12'] = df[price].ewm(span = 12).mean()
df['ema26'] = df[price].ewm(span = 26).mean()

#3.) Calculate MACD
df['MACD'] = df['ema12'] - df['ema26']
df['MACD signal'] = df['MACD'].ewm(span = 9).mean()
df.head()

#4.) Specify positions and trades
df['Positions'] = np.nan
df['Positions'] = np.where(df['MACD'] > df['MACD signal'],1,np.nan)
df['Positions'] = np.where(df['MACD'] < df['MACD signal'],-1,\
                    df['Positions'])
df['Positions'].fillna(0)
# change the first 12 data points to be 0 because we dont start trade with less that 12 data.
df['Positions'].iloc[0:12] = 0
df['Trades'] = df['Positions'].diff()
df = df.dropna()

df.to_excel('compare.xlsx')
df.head()
```

### <font color = red> Answer 1 is presented in the cell below: </font>

```python
#5.) Visualize positions
fig, axes = plt.subplots(nrows=2, ncols = 1, figsize = (18,10), sharex=True)
df[[price,'ema12','ema26']].plot(ax = axes[0], style = ['grey','red','orange'])
axes[0].set_ylabel('price')
df[['MACD','MACD signal']].plot(ax=axes[1], style = ['blue','grey'])
axes[1].set_ylabel('MACD')
bullish_condition_1 = df['Positions'] == 1
change_in_position = df['Positions'] != df['Positions'].shift(1)
bearish_condition_1 = df['Positions'] == -1

#2 plot buy positions
axes[0].plot(df[bullish_condition_1 & change_in_position].index,\
            df[bullish_condition_1 & change_in_position][price] - 2,"^", color = 'green', markersize = 10)
axes[1].plot(df[bullish_condition_1 & change_in_position].index,\
            df[bullish_condition_1 & change_in_position]['MACD'] - 0.2,"^", color = 'green', markersize = 10)

#3 plot sell position
axes[0].plot(df[bearish_condition_1 & change_in_position].index,\
            df[bearish_condition_1 & change_in_position][price] + 2,"v", color = 'red', markersize = 10)
axes[1].plot(df[bearish_condition_1 & change_in_position].index,\
            df[bearish_condition_1 & change_in_position]['MACD'] + 0.2,"v", color = 'red', markersize = 10)
fig.suptitle(ticker)
plt.show()
```

### <font color = blue> ðŸ‘‰ Question 2</font>. Letâ€™s suppose that you started from a $100,000 capital base for the given security.

### Given a $5 fixed commission fee per trade, how much in cumulative returns could you have had as a result of the MACD crossover strategy?

### Below are the lines of code that lead to an answer:

```python
#2.) approach 2

df_return2 = df[['Positions', 'Trades','return']].copy()
df_return2['strategy_return'] = df_return2['Positions'].shift(1)*df_return2['return']
commission = 5
start_value = 100000
dollar_return_dict = {}
dollar_return_dict[df_return2.index[0]] = start_value
for i in range(1, len(df_return2)):
    if df_return2['Positions'][i] == 0:
        dollar_return_dict[df_return2.index[i]] = start_value
    else:
        dollar_return_dict[df_return2.index[i]] =\
        (
            (dollar_return_dict[df_return2.index[i-1]]
             - (np.abs(df_return2['Trades'][i]) * commission))
             * (1+ df_return2['strategy_return'][i])
        )
df_return2['dollar_return'] = pd.Series(dollar_return_dict)
display(((df_return2['dollar_return'][-1] / start_value) - 1)*100)
df_return2['cumulative_net_return'] = (df_return2['dollar_return'].pct_change() +1).cumprod()
df_return2 = df_return2.dropna()
df_return2['cumulative_net_return'].plot(figsize = [18,9], c = 'black')
```

### <font color = red> Answer 2 </font>

    The answer is ___79.29_________ % .

### <font color = blue> ðŸ‘‰ Question 3</font>. Now please find the `five` worst drawdown periods over the investment horizon and provide their net drawdown in % and duration, respectively.

### <font color = green> The question asks you to find the five worst drawdown periods, beyond identifying max drawdown (MDD).

### Below are the lines of code that lead to an answer:

```python
#1.) Specify max performance
df_return2['max_gross_performance'] = df_return2['cumulative_net_return'].cummax()

#2.) Specify 5 max drawdown
df_return2['drawdown'] = df_return2['max_gross_performance'] - df_return2['cumulative_net_return']
df_return2 = df_return2.dropna()
drawdown_data = {}
for i in range (0,len(df_return2)):
    if df_return2['max_gross_performance'][i] not in drawdown_data:
        drawdown_data[df_return2['max_gross_performance'][i]] = {}
        drawdown_data[df_return2['max_gross_performance'][i]]['start_date'] = df_return2.index[i]
        drawdown_data[df_return2['max_gross_performance'][i]]['stop_date'] = df_return2.index[i]
        drawdown_data[df_return2['max_gross_performance'][i]]['max_dd'] = df_return2['drawdown'][i]
    else:
        drawdown_data[df_return2['max_gross_performance'][i]]['stop_date'] = df_return2.index[i]
        if drawdown_data[df_return2['max_gross_performance'][i]]['max_dd'] < df_return2['drawdown'][i]:
            drawdown_data[df_return2['max_gross_performance'][i]]['max_dd'] = df_return2['drawdown'][i]

worst5 = pd.DataFrame(drawdown_data).T.sort_values(by='max_dd',ascending=False).iloc[0:5]
worst5['period'] = pd.to_datetime(worst5['stop_date']) - pd.to_datetime(worst5['start_date'])
display(worst5)

#3.) Visualize
df_return2[['cumulative_net_return','max_gross_performance']].plot(figsize = [18,9], color = ['black','red'])
```

### <font color = blue> ðŸ‘‰ Question 4</font>. Within the investment horizon, please calculate annual returns for each of the three years and then find the year that performs better than average vs. less than average annual returns.

### Below are the lines of code that lead to an answer:

```python
years = [2015,2016,2017]
yearly_ret = []
for year in years:
    yearly_ret.append((df_return2.loc[str(year)]['cumulative_net_return'][-1]\
                       - df_return2.loc[str(year)]['cumulative_net_return'][0])\
                      /df_return2.loc[str(year)]['cumulative_net_return'][0])

for year, annual_return in zip(years, yearly_ret):
    print('The annual return in {} is {:.3f} %'.format(year,annual_return * 100))

print('')
print('The average return for 3 years is {:.3f} %'.format(sum(yearly_ret)/3*100))
```

###  <font color = blue> ðŸ‘‰ Question 8. </font> Which of the stocks displays the worst max drawdown, based on net drawdown % and the number of days?

### Below are the lines of code that lead to an answer:

```python
def drawdown_macd(ticker, start, end):
    #1.) import data
    data = pdr.get_data_yahoo(ticker, start = start, end = end)
    
    #2.) Calculate return and ema
    price = 'Close'
    df = data[[price]].copy()
    df['return'] = df[price].pct_change()
    df['ema12'] = df[price].ewm(span = 12).mean()
    df['ema26'] = df[price].ewm(span = 26).mean()

    #3.) Calculate MACD
    df['MACD'] = df['ema12'] - df['ema26']
    df['MACD signal'] = df['MACD'].ewm(span = 9).mean()
    df.head()

    #4.) Specify positions and trades
    df['Positions'] = np.nan
    df['Positions'] = np.where(df['MACD'] > df['MACD signal'],1,np.nan)
    df['Positions'] = np.where(df['MACD'] < df['MACD signal'],-1,\
                        df['Positions'])
    df['Positions'].fillna(0)
    # change the first 12 data points to be 0 because we dont start trade with less that 12 data.
    df['Positions'].iloc[0:12] = 0
    df['Trades'] = df['Positions'].diff()
    df = df.dropna()
    
    #5.) Calculate Return
    df_return2 = df[['Positions', 'Trades','return']].copy()
    df_return2['strategy_return'] = df_return2['Positions'].shift(1)*df_return2['return']
    commission = 5
    start_value = 100000
    dollar_return_dict = {}
    dollar_return_dict[df_return2.index[0]] = start_value
    for i in range(1, len(df_return2)):
        if df_return2['Positions'][i] == 0:
            dollar_return_dict[df_return2.index[i]] = start_value
        else:
            dollar_return_dict[df_return2.index[i]] =\
            (
                (dollar_return_dict[df_return2.index[i-1]]
                 - (np.abs(df_return2['Trades'][i]) * commission))
                 * (1+ df_return2['strategy_return'][i])
            )
    df_return2['dollar_return'] = pd.Series(dollar_return_dict)
    df_return2['cumulative_net_return'] = (df_return2['dollar_return'].pct_change() +1).cumprod()
    df_return2 = df_return2.dropna()
    
    #6.) Specify max performance
    df_return2['max_gross_performance'] = df_return2['cumulative_net_return'].cummax()

    #7.) Specify 5 max drawdown
    df_return2['drawdown'] = df_return2['max_gross_performance'] - df_return2['cumulative_net_return']
    df_return2 = df_return2.dropna()
    drawdown_data = {}
    for i in range (0,len(df_return2)):
        if df_return2['max_gross_performance'][i] not in drawdown_data:
            drawdown_data[df_return2['max_gross_performance'][i]] = {}
            drawdown_data[df_return2['max_gross_performance'][i]]['start_date'] = df_return2.index[i]
            drawdown_data[df_return2['max_gross_performance'][i]]['stop_date'] = df_return2.index[i]
            drawdown_data[df_return2['max_gross_performance'][i]]['max_dd'] = df_return2['drawdown'][i]
        else:
            drawdown_data[df_return2['max_gross_performance'][i]]['stop_date'] = df_return2.index[i]
            if drawdown_data[df_return2['max_gross_performance'][i]]['max_dd'] < df_return2['drawdown'][i]:
                drawdown_data[df_return2['max_gross_performance'][i]]['max_dd'] = df_return2['drawdown'][i]

    worst = pd.DataFrame(drawdown_data).T
    return worst
```

```python
tickers = ['MSFT','AMZN','UAL','DAL','XOM','CVX']
start = '2015-01-01'
end = '2017-12-31'

compare_drawdown = {}
for ticker in tickers:
    compare_drawdown[ticker] = drawdown_macd(ticker, start, end)
all_drawdown = pd.DataFrame({})
for key,value in compare_drawdown.items():
    value['ticker'] = key
    all_drawdown = pd.concat([all_drawdown, value])
    
all_drawdown['period'] = pd.to_datetime(all_drawdown['stop_date']) - pd.to_datetime(all_drawdown['start_date'])
display(all_drawdown.sort_values(by='max_dd',ascending = False).head(3))
display(all_drawdown.sort_values(by='period',ascending = False).head(3))
```

###  <font color = blue> ðŸ‘‰ Question 10. </font> Which of the six stocks make the current strategy investment performance? Please rank the seven stocks based on their annualized Sharpe ratio.

### Below are the lines of code that lead to an answer:

```python
def sharpe_macd(ticker, start, end):
    #1.) import data
    data = pdr.get_data_yahoo(ticker, start = start, end = end)
    
    #2.) Calculate return and ema
    price = 'Close'
    df = data[[price]].copy()
    df['return'] = df[price].pct_change()
    df['ema12'] = df[price].ewm(span = 12).mean()
    df['ema26'] = df[price].ewm(span = 26).mean()

    #3.) Calculate MACD
    df['MACD'] = df['ema12'] - df['ema26']
    df['MACD signal'] = df['MACD'].ewm(span = 9).mean()
    df.head()

    #4.) Specify positions and trades
    df['Positions'] = np.nan
    df['Positions'] = np.where(df['MACD'] > df['MACD signal'],1,np.nan)
    df['Positions'] = np.where(df['MACD'] < df['MACD signal'],-1,\
                        df['Positions'])
    df['Positions'].fillna(0)
    # change the first 12 data points to be 0 because we dont start trade with less that 12 data.
    df['Positions'].iloc[0:12] = 0
    df['Trades'] = df['Positions'].diff()
    df = df.dropna()
    
    #5.) Calculate Return
    df_return2 = df[['Positions', 'Trades','return']].copy()
    df_return2['strategy_return'] = df_return2['Positions'].shift(1)*df_return2['return']
    commission = 5
    start_value = 100000
    dollar_return_dict = {}
    dollar_return_dict[df_return2.index[0]] = start_value
    for i in range(1, len(df_return2)):
        if df_return2['Positions'][i] == 0:
            dollar_return_dict[df_return2.index[i]] = start_value
        else:
            dollar_return_dict[df_return2.index[i]] =\
            (
                (dollar_return_dict[df_return2.index[i-1]]
                 - (np.abs(df_return2['Trades'][i]) * commission))
                 * (1+ df_return2['strategy_return'][i])
            )
    df_return2['dollar_return'] = pd.Series(dollar_return_dict)
    df_return2['cumulative_net_return'] = (df_return2['dollar_return'].pct_change() +1).cumprod()
    df_return2 = df_return2.dropna()
    
    #6.) Calculate Sharpe ratio
    df_return2['daily_dollar_return'] = df_return2['dollar_return'].pct_change()
    sharpe_ratio = np.sqrt(252) * (df_return2['daily_dollar_return'].mean() / df_return2['daily_dollar_return'].std())
    
    return sharpe_ratio
```

```python
tickers = ['MSFT','AMZN','UAL','DAL','XOM','CVX']
start = '2015-01-01'
end = '2017-12-31'
compare_sharpe = {}
for ticker in tickers:
    compare_sharpe[ticker] = sharpe_macd(ticker,start,end)
pd.Series(compare_sharpe).sort_values(ascending = False)
```
